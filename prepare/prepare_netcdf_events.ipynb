```python
import xarray as xr
import pandas as pd
import os
import glob
```

```python
def search_file(file):
    print("file to search", file)
    directory = r"C:\Users\micha\Stichting Deltares\Tiaravanni Hermawan - Indonesia case for students\semarang\hydromt_data_1\to_clean\events_new"
    #print(directory)
    matching_file = glob.glob(os.path.join(directory, file))
    #print(matching_file)
    return matching_file
```

```python
# open csv
csv_path = r"C:\Users\micha\Stichting Deltares\Tiaravanni Hermawan - Indonesia case for students\semarang\hydromt_data_1\to_clean\dynamic\event_forcing.csv"
df = pd.read_csv(csv_path, sep=';')
df
```

```python
for index, row in df.iterrows():
    
    event = f"event_{row['event']}.nc"

    event_windows = row['event_windows']
    
    p_m = row['p_m']
    h_m = row['h_m']
    v_Mcm = row['v_Mcm']
    conditions = row['conditions']
    event_nc_path = search_file(event)[0]
    
    ds = xr.open_dataset(event_nc_path)
    ds.attrs["p_m"] = p_m
    ds.attrs["h_m"] = h_m
    ds.attrs["v_Mcm"] = v_Mcm
    ds.attrs["conditions"] = conditions
    modified_event_nc_path = f"C:/Users/micha/Stichting Deltares/Tiaravanni Hermawan - Indonesia case for students/semarang/hydromt_data_1/to_clean/events_with_attributes_new/{event}"
    print(modified_event_nc_path)
    ds.to_netcdf(modified_event_nc_path) 
    
```
